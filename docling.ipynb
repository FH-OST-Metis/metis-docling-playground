```python
from docling.document_converter import DocumentConverter
```

Einzelene PDF Convertieren zu Markdown

```python
converter = DocumentConverter()
result = converter.convert("pdfs/0704.0002.pdf")
print(result.document.export_to_markdown())
```

Viele convertieren zu Markdown

```python
INPUT = ["pdfs/0704.0010.pdf", "pdfs/0704.0023.pdf", "pdfs/0704.0026.pdf"]

converter = DocumentConverter()
for path in INPUT:
    result = converter.convert(path)
    print(result.document.export_to_markdown())
```
